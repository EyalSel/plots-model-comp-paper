```python
import os
import sys
import json
import time
from datetime import datetime
import numpy as np
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import pandas as pd
from collections import OrderedDict
import system_comparison_plots
sys.path.insert(1, "../optimizer/")
import single_node_profiles_cpp as snp
import seaborn as sns

%matplotlib inline
```

```python
def plot_contention(contention_path, title):
#     non_contention_path = "../results_cpp_benchmarker/single_model_profs/tf-kernel-svm/"

    contention_df = snp.create_node_profile_df(contention_path)
#     no_contention_df = snp.create_node_profile_df(non_contention_path)
#     nc_df = no_contention_df[1]
    df = contention_df[1]
    batch_sizes = df.mean_batch_size.unique()
    colors = sns.color_palette(n_colors=len(batch_sizes))
    cmap = dict(zip(batch_sizes, colors))

#     match_nc_df = nc_df[(nc_df.cloud == "aws") & (nc_df.mean_batch_size == 8)]
#     c_df = c_df.append(match_nc_df)
    fig, (ax_thru, ax_lat) = plt.subplots(ncols=2, figsize=(10,5))
#     df = df.sort_values("contention")
    for batch_size, bs_group in df.groupby(df.mean_batch_size):
        bs_group = bs_group.sort_values("contention")
        ax_thru.errorbar(bs_group.contention, bs_group.thru_stage_mean_throughput_qps, yerr=bs_group.thru_stage_std_throughput_qps, c=cmap[batch_size], label=batch_size)
        ax_lat.plot(bs_group.contention, bs_group.p99_latency, c=cmap[batch_size], label="tf-kernel-svm")
    ax_thru.set_ylim(bottom=0)
    ax_lat.set_ylim(bottom=0)
    ax_thru.set_xlabel("contention")
    ax_thru.set_ylabel("Mean throughput")
    ax_lat.set_xlabel("contention")
    ax_lat.set_ylabel("p99 latency (s)")
    ax_thru.legend(loc=0)
    ax_lat.legend(loc=0)

    fig.suptitle("TF Kernel SVM Contention Sensitivity " + title)
    plt.savefig("/Users/crankshaw/Downloads/ksvm_batch_size_sweep_prof_vs_contention.png")
    plt.show()
```

```python
plot_contention("../results_cpp_benchmarker/e2e_results/image_driver_1/single_model_profiles_contention_sensitivity/tf-kernel-svm-SMP-contention-batch-size-sweep/", "Effects of Batch Size on Contention Slowdown")
```
